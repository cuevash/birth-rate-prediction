### Extracting and trasnforming

Now we select the last year for which we have data , the 2021, and extract the features selected from the different dataframes. Then we merge all the dataframes in a single one so we can carry out the calculations.

On this data frame we are going to fit a simple linear regression and see how good it fits the values we have.

```python
YEAR_IN_STUDY = "2019"
```


#### Preparing gender dataframe


```python
#
GENDER_INDICATORS = ["SL.TLF.CACT.FM.ZS", "SE.SCH.LIFE.FE"]
```

```python
# Selecting data for indicators
df_gender_selected_indicators = df_gender_indicators[
    df_gender_indicators["Indicator Code"].isin(GENDER_INDICATORS)
]

df_gender_selected_features = df_gender_selected_indicators.loc[
    :,
    ["Country Name", "Country Code", "Indicator Name", "Indicator Code", YEAR_IN_STUDY],
]
```

```python
df_gender_selected_features
```

```python
# Remove countries with nan values on features

countries_with_NaN = df_gender_selected_features[
    np.isnan(df_gender_selected_features[YEAR_IN_STUDY])
]["Country Code"].compute()

df_gender_selected_with_data = df_gender_selected_features[
    ~df_gender_selected_features["Country Code"].isin(countries_with_NaN)
]

df_gender_selected_with_data
```

```python
#
WORLD_INDICATORS = ["SP.DYN.TFRT.IN", "NY.GNP.PCAP.CD", "SP.POP.TOTL"]
```

```python
# Selecting data for indicators
df_world_selected_indicators = df_world_indicators[
    df_world_indicators["Indicator Code"].isin(WORLD_INDICATORS)
]

df_world_selected_features = df_world_selected_indicators.loc[
    :,
    ["Country Name", "Country Code", "Indicator Name", "Indicator Code", YEAR_IN_STUDY],
]
```

```python
df_world_selected_features
```

```python
# Remove countries with nan values on features

countries_with_NaN = df_world_selected_features[
    np.isnan(df_world_selected_features[YEAR_IN_STUDY])
]["Country Code"].compute()

df_world_selected_with_data = df_world_selected_features[
    ~df_world_selected_features["Country Code"].isin(countries_with_NaN)
]

df_world_selected_with_data
```

```python
countries_in_all_dfs = set(df_gender_selected_with_data["Country Code"]) & set(
    df_world_selected_with_data["Country Code"]
)
countries_in_all_dfs
```

```python
df_merged = dd.concat(
    [
        df_gender_selected_features[
            df_gender_selected_features["Country Code"].isin(countries_in_all_dfs)
        ],
        df_world_selected_features[
            df_world_selected_features["Country Code"].isin(countries_in_all_dfs)
        ],
    ]
)

df_merged
```

```python
# prepare df for pivoting
df_merged_pre_pivot = df_merged.categorize(columns=["Indicator Code"])

df_merged_pivot = df_merged_pre_pivot.pivot_table(
    index="Country Code", columns="Indicator Code", values="2019"
)

df_merged_final = df_merged_pivot.compute()
df_merged_final
```

```python
df_merged_final

# Renaming columns & reordering columns
df_fertility_regresion_data = (
    df_merged_final.rename(
        columns={
            "NY.GNP.PCAP.CD": "gni_per_capita_atlas_method",
            "SE.SCH.LIFE.FE": "female_expected_years_schooling",
            "SL.TLF.CACT.FM.ZS": "ratio_female_to_male_labour_participation",
            "SP.POP.TOTL": "total_population",
            "SP.DYN.TFRT.IN": "fertility_rate",
        }
    )
    .rename_axis(index="country_code")
    .reindex(
        columns=[
            "gni_per_capita_atlas_method",
            "female_expected_years_schooling",
            "ratio_female_to_male_labour_participation",
            "total_population",
            "fertility_rate",
        ]
    )
)

df_fertility_regresion_data
```

### Visualizing some exploratory graphs.

```python
# Ratio Func
def ratio_scatter(x_col, y_col, title, x_axis_label, y_axis_label, data):

    p = figure(
        title=title,
        x_axis_label=x_axis_label,
        y_axis_label=y_axis_label,
        background_fill_color="#fafafa",
    )

    # scatter plot

    p.scatter(x_col, y_col, source=data)

    # show the results
    show(p)
```

```python
# Ratio with the variables independently.

ratio_scatter(
    "gni_per_capita_atlas_method",
    "fertility_rate",
    data=df_fertility_regresion_data,
    title="Ratio",
    x_axis_label="gni_per_capita_atlas_method",
    y_axis_label="fertility_rate",
)

ratio_scatter(
    "female_expected_years_schooling",
    "fertility_rate",
    data=df_fertility_regresion_data,
    title="Ratio",
    x_axis_label="female_expected_years_schooling",
    y_axis_label="fertility_rate",
)


ratio_scatter(
    "ratio_female_to_male_labour_participation",
    "fertility_rate",
    data=df_fertility_regresion_data,
    title="Ratio",
    x_axis_label="ratio_female_to_male_labour_participation",
    y_axis_label="fertility_rate",
)

ratio_scatter(
    "total_population",
    "fertility_rate",
    data=df_fertility_regresion_data,
    title="Ratio",
    x_axis_label="total_population",
    y_axis_label="fertility_rate",
)
```
